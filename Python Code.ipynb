```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
train_df
```

```python
test_df
```

```python
x_train = train_df[train_df.columns[1:-1]]
x_train = x_train.dropna()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=372)

x_train_pca = pca.fit_transform(x_train)

x_train_pca
```

```python
x_train_pca.shape
```

```python
y_train2 = train_df[train_df.columns[-1]]
y_train2
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
print(len(x_train_pca[0])) #Before LDA

lda.fit(x_train_pca, y_train2)
x_train_lda = lda.transform(x_train_pca)

len(x_train_lda[0]) #After LDA
```

```python
from sklearn.neighbors import LocalOutlierFactor

lof=LocalOutlierFactor(n_neighbors=5, contamination=0.05)
y_pred=lof.fit_predict(x_train_lda)

outliers=np.where(y_pred==-1)[0]
outliers
```

```python
x_train_lda_outRemoved=np.delete(x_train_lda, outliers, axis=0)
```

```python
y_train=[]
for i in range(1216):
    if(i not in outliers):
        y_train.append(y_train2[i])
```

```python
y_train=np.array(y_train)
y_train2=y_train
```

```python
from sklearn.cluster import KMeans

kmeans= KMeans(n_clusters=1, n_init=10)
kmeans.fit(x_train_lda_outRemoved)

x_train_lda_outRemoved=np.column_stack((x_train_lda_outRemoved, kmeans.labels_))
```

```python
x_train_lda_outRemoved=x_train_lda_outRemoved[ :, :-1]
x_train_lda_outRemoved.shape
```

```python
y_train2.shape
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression(max_iter=3000)
lor.fit(x_train_lda_outRemoved,y_train2)
```

```python
x_test = test_df[test_df.columns[1:]]
x_test
```

```python
x_test_pca = pca.transform(x_test)
print(len(x_test_pca[0]))
```

```python
x_test_lda=lda.transform(x_test_pca)
x_test_lda.shape
```

```python
x_train_lda=np.column_stack((x_test_lda, kmeans.predict(x_test_lda)))
```

```python
y_pred_labels = lor.predict(x_test_lda)
y_pred_labels
```

```python
y_pred_labels_df = pd.DataFrame(y_pred_labels, columns=['category'])
y_pred_labels_df
```

```python
y_pred_labels_df.to_csv('predictionsLogisticRegression.csv', sep=',', index=True)
```
